```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
x = pd.DataFrame({
    "something": [1, 2, 3]
})

y = [1, 3, 6, 10]

chart1 = plt.plot(x);
chart2 = plt.plot(y);
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie(y, labels= ["Mar", "Apr", "Jul", "Jun"], explode=[0.1, 0.1, 0.1, 0.1]);
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
x = np.random.randn(50)
y = np.random.randn(50)

scatter = plt.scatter(x, y, alpha= 0.4, s= 50);
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
x = np.random.randn(50)
y = np.random.randn(50)

scatter = plt.scatter(x, y, alpha= 0.4, s= 50);
plt.title("something", loc="left", c="red", fontsize= 16);
```

## Q5: Create three line charts with Legends at left and colors

```python
plt.plot(x);
plt.plot(x+1);
plt.plot(x-1);
plt.legend(["blue", "orange", "green"]);
```

## Q6: Create subplots with three charts and figsize=20,10

```python
plt.subplot(2,2,1);
plt.plot(x);
plt.subplot(2,2,2);
plt.plot(x+1);
plt.subplot(2,2,3);
plt.plot(x-1);
```
